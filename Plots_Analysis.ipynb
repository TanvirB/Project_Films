```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Plot popular movie genre as pie chart
movie_genre = pd.read_csv("Resources/US_movies_Genre.csv")
movie_genre.drop(columns = ["Unnamed: 0"],inplace=True)
movie_genre.head()
```

```python
movie_genre.sort_values(by=["Genre Counts"], ascending=False, inplace= True)
movie_genre_top = movie_genre[:10].copy()
new_row = pd.DataFrame(data = {"Genre" : ["Others"],
                               "Genre Counts" : [movie_genre["Genre Counts"][10:].sum()]})
movie_genre_top = pd.concat([movie_genre_top, new_row])

labels = movie_genre_top["Genre"]
size   = movie_genre_top["Genre Counts"]
color = plt.cm.jet(np.linspace(0,1,len(size)+5))
explode = []; explode.extend(np.zeros(len(size)-1)); explode.extend([0.2])


fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))
patches, texts, autotexts = plt.pie(size, explode=explode, labels=labels, colors=color[3:],
        autopct="%1.1f%%", shadow=False, startangle=50, radius=1.2)
plt.setp(autotexts, size=15, weight="bold")
plt.setp(texts, size=15, weight="bold")

# plt.savefig("Figures/Movie_Genre_PieChart.png")

plt.show()
```
