```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Plot popular movie genre as pie chart
movie_genre = pd.read_csv("Resources/US_movies_Genre.csv")
movie_genre.drop(columns = ["Unnamed: 0"], inplace=True)
movie_genre.head()
```

```python
labels = movie_genre["Genre"]
size   = movie_genre["Genre Counts"]
color = plt.cm.rainbow(np.linspace(0,1,len(size)))
explode = np.zeros(len(size)) 

plt.figure(figsize=(8,8))
plt.pie(size, explode=explode, labels=labels, colors=color,
        autopct="%1.1f%%", shadow=False, startangle=140)
plt.show()
```

```python
# Plot popular Production as histogram
movie_production = pd.read_csv("Resources/US_movies_Production.csv")
movie_production.drop(columns = ["Unnamed: 0"], inplace=True)
movie_production.head()
```

```python
movie_production.sort_values(by=["Production Counts"], ascending=False, inplace= True)
movie_production.head()
```

```python
plt.figure(figsize=(15,5))
bar = plt.bar(movie_production["Production"][1:11], 
              movie_production["Production Counts"][1:11],
              color='b', alpha=0.5)
plt.ylabel("Production Counts")
plt.xticks(rotation=30)
plt.show()
```

```python
# Plot popular actors as histogram
movie_actors = pd.read_csv("Resources/US_movies_Actors.csv")
movie_actors.drop(columns = ["Unnamed: 0"], inplace=True)
movie_actors.head()
```

```python
movie_actors.sort_values(by=["Actor Counts"], ascending=False, inplace=True)
movie_actors.head()
```

```python
plt.figure(figsize=(15,5))
bar = plt.bar(movie_actors["Actor Names"][0:30], 
              movie_actors["Actor Counts"][0:30],
              color='b', alpha=0.5)
plt.ylabel("Actor Counts")
plt.xticks(rotation=30)
plt.show()
```
